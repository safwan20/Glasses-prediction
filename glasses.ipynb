```python
#image processing
import cv2
from keras.preprocessing import image

#graph plot
import matplotlib.pyplot as plt

#data aumentation
from keras.preprocessing.image import ImageDataGenerator

#reading csv
import pandas as pd

#data manupilation
import numpy as np

#model preparation
import tensorflow as tf
from keras.layers import Conv2D, MaxPooling2D , Flatten, Dense, Dropout
from keras.models import Sequential
from tensorflow import keras
```

```python
dataset_path = ''
data = cv2.imread(dataset_path)
plt.imshow(data)
```

```python
image_gen = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.1,
    height_shift_range=0.1,
    rescale=1/255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
train_image_gen = image_gen.flow_from_directory('/kaggle/input/datata/datasetss/samples/trains/',color_mode ='grayscale',target_size=(48,48),batch_size=32,class_mode='binary')
```

```python
plt.imshow(image_gen.random_transform(data))
```

```python
classifier = Sequential()
classifier.add(Conv2D(32,(3,3),padding="same",input_shape = (48,48,1),activation='relu'))
classifier.add(MaxPooling2D(pool_size=(2,2)))
classifier.add(Dropout(0.2))
classifier.add(Conv2D(32,(3,3),padding="same",activation='relu'))
classifier.add(MaxPooling2D(pool_size=(2,2)))
classifier.add(Dropout(0.2))
classifier.add(Conv2D(32,(3,3),padding="same",activation='relu'))
classifier.add(MaxPooling2D(pool_size=(2,2)))
classifier.add(Flatten())
classifier.add(Dense(units=128,activation='relu'))
classifier.add(Dropout(0.2))
classifier.add(Dense(units=1,activation='sigmoid'))
classifier.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
results = classifier.fit_generator(
    train_image_gen,
    epochs=10,
)
```

```python
test_file = ''
data_test = cv2.imread(test_file)
plt.imshow(data_test)
```

```python
loaded_image = image.load_img(test_file,target_size=(48,48),color_mode='grayscale')
loaded_image_array = image.img_to_array(loaded_image)
lets_predict_this = np.expand_dims(loaded_image_array,axis=0)
lets_predict_this = lets_predict_this/255
classifier.predict_classes(lets_predict_this)
```
